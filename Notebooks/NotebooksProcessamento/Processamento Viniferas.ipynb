```python
import requests
import pandas as pd
import json

# URLs dos arquivos JSON no GitHub
url_viniferas = 'https://raw.githubusercontent.com/Data-Analitycs-Pos-Tech-Fiap/DadosVinifera/refs/heads/main/bases_de_dados/dados_processamento/viniferas.json'
url_americanas = 'https://raw.githubusercontent.com/Data-Analitycs-Pos-Tech-Fiap/DadosVinifera/refs/heads/main/bases_de_dados/dados_processamento/americanas.json'
url_mesa = 'https://raw.githubusercontent.com/Data-Analitycs-Pos-Tech-Fiap/DadosVinifera/refs/heads/main/bases_de_dados/dados_processamento/mesa.json'

# Dicionário para armazenar os DataFrames
dataframes = {}

# Lista de URLs para iteração
urls = {
    'viniferas': url_viniferas,
    'americanas': url_americanas,
    'mesa': url_mesa,
}

# Fazer a requisição HTTP para cada URL
for key, url in urls.items():
    response = requests.get(url)

    # Verificar se a requisição foi bem-sucedida
    if response.status_code == 200:
        # Tentativa de decodificação do JSON
        try:
            # Capturando o texto da resposta
            content = response.text

            # Transformando o conteúdo em uma lista de objetos JSON
            json_data = [json.loads(line) for line in content.strip().splitlines()]

            # Transformar o JSON em um DataFrame (estrutura tabular)
            dataframes[key] = pd.json_normalize(json_data)

            # Exibir as primeiras linhas do DataFrame para visualização
            print(f"DataFrame para {key}:")
            print(dataframes[key].head())
        except ValueError as e:
            print(f"Erro ao decodificar JSON para {key}: {e}")
            print("Conteúdo da resposta:")
            print(response.text)  # Imprime o conteúdo bruto da resposta
    else:
        print(f"Erro ao acessar {key}. Código de status: {response.status_code}")
```

```python
# Acessando cada DataFrame
viniferas = dataframes['viniferas']
americanas = dataframes['americanas']
mesa = dataframes['mesa']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selecionando os anos relevantes para calcular o total por ano
anos = viniferas.columns[1:15]
total_por_ano = viniferas[anos].sum() / 1_000_000  # Convertendo para milhões

# Gráfico de linha para o total de custos ao longo dos anos
plt.figure(figsize=(10, 5))
plt.plot(total_por_ano.index, total_por_ano.values, marker='o')
plt.title('Total de Custos de Processamento de Vinhos ao Longo dos Anos (em Milhões)')
plt.xlabel('Ano')
plt.ylabel('Custo Total (R$ Milhões)')  # Ajuste no rótulo para refletir a unidade
plt.grid()
plt.xticks(rotation=45)  # Opcional: rotacionar os rótulos do eixo x
plt.show()

```

```python
# Transpondo o DataFrame para facilitar a visualização
viniferas_melted = viniferas.melt(id_vars=['Tipo_de_uva'], var_name='Ano', value_name='Custo')

# Calculando o custo médio por tipo de uva
custo_medio = viniferas_melted.groupby('Tipo_de_uva')['Custo'].mean().reset_index()

# Obtendo os 5 tipos de uva com os maiores custos
top5_custos = custo_medio.nlargest(5, 'Custo')

# Filtrando os dados para incluir apenas os tipos de uva que estão entre os 5 maiores custos
viniferas_top5 = viniferas_melted[viniferas_melted['Tipo_de_uva'].isin(top5_custos['Tipo_de_uva'])]

# Análise de Tendências de Custos para os 5 tipos de uva com maiores custos
plt.figure(figsize=(12, 6))

# Criando o gráfico de linha
ax = sns.lineplot(data=viniferas_top5, x='Ano', y='Custo', hue='Tipo_de_uva', marker='o')

# Adicionando linhas de referência no fundo (pontilhadas e claras)
for y in range(0, int(viniferas_top5['Custo'].max()), 5000000):  # Ajuste o intervalo conforme necessário
    ax.axhline(y=y, color='lightgray', linestyle='--', linewidth=0.8)

# Adicionando linhas verticais pontilhadas para cada ano
for ano in viniferas_top5['Ano'].unique():
    ax.axvline(x=ano, color='lightgray', linestyle='--', linewidth=0.8)

# Título e rótulos
plt.title('Tendência de Custos de Processamento por Tipo de Uva (Top 5 Custos)')
plt.xticks(rotation=45)
plt.ylabel('Custo de Processamento')
plt.legend(title='Tipo de Uva')

plt.show()

```

```python
# Transpondo o DataFrame para facilitar a visualização
viniferas_melted = viniferas.melt(id_vars=['Tipo_de_uva'], var_name='Ano', value_name='Custo')

# Calculando o custo médio por tipo de uva
custo_medio = viniferas_melted.groupby('Tipo_de_uva')['Custo'].mean().reset_index()

# Obtendo os 5 tipos de uva com os maiores custos
top5_custos = custo_medio.nlargest(5, 'Custo')['Tipo_de_uva']

# Filtrando o DataFrame original para apenas os 5 tipos de uva com maiores custos
heatmap_data = viniferas[viniferas['Tipo_de_uva'].isin(top5_custos)].set_index('Tipo_de_uva').T

# Criando a matriz de calor
plt.figure(figsize=(12, 6))

# Formatando os dados para exibir em milhões
formatted_data = heatmap_data / 1_000_000

# Criando a matriz de calor
sns.heatmap(formatted_data, annot=formatted_data.applymap(lambda x: f'{x:.2f} M'), cmap='YlGnBu', fmt='')

plt.title('Matriz de Calor dos Custos de Processamento dos 5 Tipos de Uva com Maior Custo')
plt.ylabel('Ano')
plt.xlabel('Tipo de Uva')

# Rotacionando os rótulos do eixo x (Tipos de Uva) para ficar deitado
plt.xticks(rotation=45, ha='right')

# Rotacionando os rótulos do eixo y (Anos) para ficar na horizontal
plt.yticks(rotation=0)

# Ajustando o espaçamento entre os rótulos do eixo y
plt.subplots_adjust(left=0.15)  # Ajuste conforme necessário

plt.show()
```
