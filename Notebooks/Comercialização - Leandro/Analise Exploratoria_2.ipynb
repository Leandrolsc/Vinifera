É interessante utilizar a historia traçando das abas de Processamento > Produção > Exportação

Dessa forma da pra realizar uma história do começo ao destino. Se atentar de pegar somente vinhos.

```python
import pandas as pd
```

```python
#Leitura dos csv
ProcessaMesa = pd.read_csv('/content/ProcessaMesa.csv',sep='\t')
ProcessaAmericanas = pd.read_csv('/content/ProcessaAmericanas.csv',sep='\t')
ProcessaSemclass = pd.read_csv('/content/ProcessaSemclass.csv',sep='\t')
ProcessaViniferas = pd.read_csv('/content/ProcessaViniferas.csv',sep=';')
Producao = pd.read_csv('/content/Producao.csv',sep=';')
ExpVinho = pd.read_csv('/content/ExpVinho.csv',sep=';')
```

```python
#Preparando as tabelas para o Pivot
ExpVinho = ExpVinho.drop(columns=['id'], errors='ignore') #tratamento a parte
Producao = Producao.drop(columns=['id'], errors='ignore')
ProcessaViniferas = ProcessaViniferas.drop(columns=['id'], errors='ignore')
ProcessaSemclass = ProcessaSemclass.drop(columns=['id'], errors='ignore')
ProcessaAmericanas = ProcessaAmericanas.drop(columns=['id'], errors='ignore')
ProcessaMesa = ProcessaMesa.drop(columns=['id'], errors='ignore')
```

```python
#Pivot das tabelas
Producao = pd.melt(Producao,id_vars=['control','produto'],var_name='Ano',value_name='Quantidade')
ProcessaViniferas = pd.melt(ProcessaViniferas,id_vars=['control','cultivar'],var_name='Ano',value_name='Quantidade')
ProcessaSemclass = pd.melt(ProcessaSemclass,id_vars=['control','cultivar'],var_name='Ano',value_name='Quantidade')
ProcessaAmericanas = pd.melt(ProcessaAmericanas,id_vars=['control','cultivar'],var_name='Ano',value_name='Quantidade')
ProcessaMesa = pd.melt(ProcessaMesa,id_vars=['control','cultivar'],var_name='Ano',value_name='Quantidade')
```

```python
#Preparando a ExpVinho
#Onde coluna com .1 = US$
#Onde Coluna sem .1 = Kg

#Pegando colunas
ExpVinhoKg = ExpVinho[['Id','País','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']]
#Pivot como na celula anterior
ExpVinhoKg = pd.melt(ExpVinhoKg,id_vars=['Id','País'],var_name='Ano',value_name='Quantidade')

#Pegando colunas e renomeando
ExpVinhoDolar = ExpVinho[['Id','País','2009.1','2010.1','2011.1','2012.1','2013.1','2014.1','2015.1','2016.1','2017.1','2018.1','2019.1','2020.1','2021.1','2022.1','2023.1']]
ExpVinhoDolar.columns = ['Id','País','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
#Pivot como na celula anterior
ExpVinhoDolar = pd.melt(ExpVinhoDolar,id_vars=['Id','País'],var_name='Ano',value_name='Quantidade')
```

```python
#Transformando as colunas Ano em Inteiro para filtrar de 2009 em diante
Producao['Ano'] = Producao['Ano'].astype(int)
Producao = Producao[Producao['Ano'] >= 2009]
ProcessaViniferas['Ano'] = ProcessaViniferas['Ano'].astype(int)
ProcessaViniferas = ProcessaViniferas[ProcessaViniferas['Ano'] >= 2009]
ProcessaSemclass['Ano'] = ProcessaSemclass['Ano'].astype(int)
ProcessaSemclass = ProcessaSemclass[ProcessaSemclass['Ano'] >= 2009]
ProcessaAmericanas['Ano'] = ProcessaAmericanas['Ano'].astype(int)
ProcessaAmericanas = ProcessaAmericanas[ProcessaAmericanas['Ano'] >= 2009]
ProcessaMesa['Ano'] = ProcessaMesa['Ano'].astype(int)
ProcessaMesa = ProcessaMesa[ProcessaMesa['Ano'] >= 2009]
```

```python
#Tornando todos os dados da coluna control minusculos
#pegando todos os dados onde a coluna control que contenha a palavra vinho
Producao['control'] = Producao['control'].str.lower()
Producao = Producao[Producao['control'].str.contains('vinho')]
```

```python
Producao[Producao['Ano']==2023]
```

```python
Quantidade_Ano = Producao.groupby(['produto','Ano'])['Quantidade'].sum()
SomaTotal = Producao['Quantidade'].sum()
print("Quantidade Total Produzida: " + f'{SomaTotal:,.0f}'.replace(',', '.'))
print(Quantidade_Ano)
```

```python
Quantidade_Ano = Quantidade_Ano.reset_index(name='Quantidade')
```

```python
import altair as alt

# Exemplo de DataFrame agregado_processa já criado
# Criar o gráfico de linhas com uma paleta de cores personalizada
#
chart = alt.Chart(Quantidade_Ano).mark_line().encode(
    x='Ano:O',  # 'O' indica que o Ano é tratado como ordinal
    y='Quantidade:Q',
    color=alt.Color('produto:N', scale=alt.Scale(scheme='set1'))  # Alterar a paleta de cores
).properties(
    width=800,  # Definir a largura do gráfico
    height=500,  # Definir a altura do gráfico
    title='Quantidade por Ano por produto'
)


# Exibir o gráfico
chart
```

```python
#Transformando a coluna quantidade em processos viniferas em float
ProcessaViniferas['Quantidade'] = ProcessaViniferas['Quantidade'].str.replace(',', '.').replace('nd',0).replace('*',0).astype(float)
ProcessaViniferas[ProcessaViniferas['Ano'] ==2023]
```

```python
#Transformando a coluna quantidade em processos sem classificação em float
ProcessaSemclass['Quantidade'] = ProcessaSemclass['Quantidade'].replace(',', '.').replace('nd',0).replace('*',0).astype(float).fillna(0)
ProcessaSemclass[ProcessaSemclass['Ano'] ==2023]
```

```python
#Transformando a coluna quantidade em processa americanas em float
ProcessaAmericanas['Quantidade'] = ProcessaAmericanas['Quantidade'].replace(',', '.').replace('nd',0).replace('*',0).replace('+',0).astype(float).fillna(0)
ProcessaAmericanas[ProcessaAmericanas['Ano'] ==2023]
```

```python
#Processa mesa
ProcessaMesa['Quantidade'] = ProcessaMesa['Quantidade'].replace('nd',0).replace('*',0).astype(float).fillna(0)
ProcessaMesa[ProcessaMesa['Ano'] ==2023]
```

```python
#Juntando todos os dataframes do Processamento
Processa = pd.concat([ProcessaViniferas, ProcessaSemclass,ProcessaAmericanas,ProcessaMesa], ignore_index=True)
```

```python
Processa = Processa[~Processa['control'].str.contains('_')]
Processa['Quantidade'] = Processa['Quantidade'].map('{:.0f}'.format)
Processa['Quantidade'] = Processa['Quantidade'].astype(float)
Processa[Processa['Ano'] ==2023]
```

```python
#Agregar  por control, cultivar  e ano, somando a quantidade
agregado_processa = Processa.groupby(['cultivar', 'Ano'])['Quantidade'].sum().reset_index()
agregado_processa.head()
```

```python
import altair as alt

# Exemplo de DataFrame agregado_processa já criado
# Criar o gráfico de linhas com uma paleta de cores personalizada
chart = alt.Chart(agregado_processa).mark_line().encode(
    x='Ano:O',  # 'O' indica que o Ano é tratado como ordinal
    y='Quantidade:Q',
    color=alt.Color('cultivar:N', scale=alt.Scale(scheme='set1'))  # Alterar a paleta de cores
).properties(
    width=800,  # Definir a largura do gráfico
    height=500,  # Definir a altura do gráfico
    title='Quantidade por Ano para cada Cultivar'
)


# Exibir o gráfico
chart
```

```python
#Pegando todos os dados onde o control não tenha o underscore e filtrando os dados de quantidade deixando todos como inteiro
Processa = Processa[~Processa['control'].str.contains('_')]
Processa['Quantidade'] = pd.to_numeric(Processa['Quantidade'], errors='coerce').fillna(0).astype(int)
```
